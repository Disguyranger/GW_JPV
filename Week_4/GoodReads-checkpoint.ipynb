```python
# Import Dependencies
import pandas as pd
```

```python
# Make a reference to the books.csv file path
fileOne = "Resources/books.csv"
# Import the books.csv file as a DataFrame
fileOneDataFrame = pd.read_csv(fileOne)
fileOneDataFrame
```

```python
# Remove unecessary columns from the DataFrame and save the new DataFrame
changeDataFrame = fileOneDataFrame[["isbn","original_publication_year","original_title", "authors", "ratings_count","ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"]]
fileOneDataFrame = changeDataFrame
fileOneDataFrame
# Only keep: "isbn", "original_publication_year", "original_title", "authors",
# "ratings_1", "ratings_2", "ratings_3", "ratings_4", "ratings_5"
```

```python
# Rename the headers to be more explanatory
renamed_df = fileOneDataFrame.rename(columns={"isbn": "ISBN", "original_publication_year" : "Publication Year", 
                                              "original_title" : "Title", "authors" : "Author", 
                                              "ratings_count" : "Total Ratings", "ratings_1" : "Africa",
                                              "ratings_2" : "UK", "ratings_3" : "Asia",
                                              "ratings_4" : "Korea", "ratings_5" : "USA"
                                             })
renamed_df
```

```python
# Push the remade DataFrame to a new CSV file
renamed_df.to_csv('Resources/FileOne.csv', index = False, header = True)
```
